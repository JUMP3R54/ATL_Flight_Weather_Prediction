**Filter the national flight dataset doen to only ATL Flights**

The entire Flight dataset was way to big so we will only focus on the section relating to ATL airport and filter out all other airports

```python
import pandas as pd

input_file = "/content/drive/MyDrive/flight_data_2024.csv"
output_file = "/content/drive/MyDrive/atl_flights_2024.csv"

wanted_cols = [
    "year",
    "month",
    "day_of_month",
    "day_of_week",
    "fl_date",
    "op_unique_carrier",
    "op_carrier_fl_num",
    "origin",
    "origin_city_name",
    "origin_state_nm",
    "dest",
    "dest_city_name",
    "dest_state_nm",
    "crs_dep_time",
    "dep_time",
    "dep_delay",
    "taxi_out",
    "wheels_off",
    "wheels_on",
    "taxi_in",
    "crs_arr_time",
    "arr_time",
    "arr_delay",
    "cancelled",
    "cancellation_code",
    "diverted",
    "crs_elapsed_time",
    "actual_elapsed_time",
    "air_time",
    "distance",
    "carrier_delay",
    "weather_delay",
    "nas_delay",
    "security_delay",
    "late_aircraft_delay"
]

first_chunk = True
chunk_num = 0

for chunk in pd.read_csv(input_file, chunksize=100000):
    chunk_num += 1

    cols = [c for c in wanted_cols if c in chunk.columns]
    chunk = chunk[cols]

    atl_chunk = chunk[(chunk["origin"] == "ATL") | (chunk["dest"] == "ATL")]

    atl_chunk.to_csv(
        output_file,
        mode="w" if first_chunk else "a",
        index=False,
        header=first_chunk
    )

    first_chunk = False
    print(f"Finished chunk {chunk_num}, kept {len(atl_chunk)} rows")

print("Done. Saved ATL-only file to:", output_file)
```

**Clean ATL Flight File**

taking the ATL only flights we are converting the cols into the proper data types, making sure missing values are delt with and removing duplications of data

```python
#CLEAN THE DATA
import pandas as pd
import numpy as np

input_file = "/content/drive/MyDrive/atl_flights_2024.csv"
output_file = "/content/drive/MyDrive/atl_flights_2024_cleaned.csv"

df = pd.read_csv(input_file, low_memory=False)

# Basic date cleanup
df["fl_date"] = pd.to_datetime(df["fl_date"], errors="coerce")

# Remove duplicates
df = df.drop_duplicates()

# Convert numeric columns
numeric_cols = [
    "year", "month", "day_of_month", "day_of_week",
    "op_carrier_fl_num",
    "crs_dep_time", "dep_time", "dep_delay",
    "taxi_out", "wheels_off", "wheels_on", "taxi_in",
    "crs_arr_time", "arr_time", "arr_delay",
    "cancelled", "diverted",
    "crs_elapsed_time", "actual_elapsed_time", "air_time", "distance",
    "carrier_delay", "weather_delay", "nas_delay",
    "security_delay", "late_aircraft_delay"
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Clean text columns
text_cols = [
    "op_unique_carrier", "origin", "origin_city_name", "origin_state_nm",
    "dest", "dest_city_name", "dest_state_nm", "cancellation_code"
]

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()

# Fix placeholder text values
if "cancellation_code" in df.columns:
    df.loc[df["cancellation_code"].isin(["nan", "None", "NaN"]), "cancellation_code"] = np.nan

# Fill some basic flag columns
if "cancelled" in df.columns:
    df["cancelled"] = df["cancelled"].fillna(0).astype(int)

if "diverted" in df.columns:
    df["diverted"] = df["diverted"].fillna(0).astype(int)

# Fill delay breakdown columns only
delay_breakdown_cols = [
    "carrier_delay", "weather_delay", "nas_delay",
    "security_delay", "late_aircraft_delay"
]

for col in delay_breakdown_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# Drop rows missing core identifiers
df = df.dropna(subset=["fl_date", "origin", "dest", "op_unique_carrier"])

df.to_csv(output_file, index=False)

print("Done. Saved cleaned file to:", output_file)
print("Rows:", len(df))
print("Columns:", len(df.columns))
print(df.head())
```

**Prepare/Load Weather Data**

Import the Atlanta weather dataset, convert the month names into numeric values for proper joining, and laod the weahter data into SQLite.  This is so the weather table matches the flight table more cleanly when joined

```python
#IMPORT THE WEATHER DATASET AND CREATE NUMERIC MONTH VALUES

weather = pd.read_csv("/content/drive/MyDrive/atlanta_2024_weather_daily.csv")

change_month = {
    "jan": 1, "feb": 2, "mar": 3, "apr": 4, "may": 5, "jun": 6, "jul": 7, "aug": 8, "sep": 9, "oct": 10, "nov": 11, "dec": 12
}
weather['Month_num'] = weather["Month"].str.lower().map(change_month)
weather = weather.drop(columns=["Month"])

#TURN THE WEATHER DATA INTO A SQL TABLE
conn = sqlite3.connect("atl_project.db")

weather.to_sql("weather_atl", conn, if_exists="replace", index=False)
print("weather_atl table made")

pd.read_sql_query("SELECT * FROM weather_atl LIMIT 5;", conn)
```

**Join Flight and Weather Tables**

Here we are joining the cleaned flight table and the weather table using month and day.  Once this is done each flight row also has the corresponding weather info for that same day in ATL.

```python
#CONNECT THE TWO TABLES TOGETHER
weather.to_sql("weather_atl", conn, if_exists="replace", index=False)
flights.to_sql("flights_atl", conn, if_exists="replace", index=False)
query = """
CREATE TABLE IF NOT EXISTS joined_atl AS
SELECT *
FROM flights_atl flights
JOIN weather_atl weather
  ON flights.day_of_month = weather.Day
  AND flights.month = weather.Month_num;
"""
cursor = conn.cursor()
cursor.execute(query)
conn.commit()

pd.read_sql_query("SELECT * FROM joined_atl LIMIT 5;", conn)
```

**Helper Cols for Testing**

**Values for Helper Cols**

This section updates the helper cols we created above.  It makrs flights with delays over 15 mins, and creates a simple weather severity flag case (will be used to predict if weather will delay flights)

```python
#Create severity flags
pd.read_sql_query("SELECT * FROM joined_atl LIMIT 5;", conn)

cursor.execute("""
    UPDATE joined_atl
    SET
      dep_delayed_15 = CASE
        WHEN cancelled = 0 AND dep_delay > 15 THEN 1
        ELSE 0
      END,
      arr_delayed_15 = CASE
        WHEN cancelled = 0 AND arr_delay > 15 THEN 1
        ELSE 0
      END,
      weather_severity_flag = CASE
        WHEN Precip_Total_in > 1.25 OR Wind_Avg_mph >= 15 THEN 1
        ELSE 0
      END,
      sched_dep_hour = CAST(crs_dep_time / 100 AS INTEGER);
""")
conn.commit()
print("This worked *thumbs up* p2")
pd.read_sql_query("SELECT * FROM joined_atl LIMIT 100;", conn)
```

**SQL summary: airlines with the highest average arrival delay**

This quert is one of our first checks.  It helps show which ariliens tend to have the highest average arrival delays in dataset.

```python
#See what airlines are the most likely to be delayed
pd.read_sql_query("""
SELECT
    op_unique_carrier,
    AVG(arr_delay) AS avg_arr_delay
FROM joined_atl
WHERE cancelled = 0
GROUP BY op_unique_carrier
ORDER BY avg_arr_delay DESC;
""", conn)
```

**SQL summary: weather severity and disruption**

Looks at how the weather severity flag lines up with the delay and cancellation rates.  It give us an idea of whether weather conditions appear to matter in the joined data.

```python
#See what % of flights are cancelled due to having weather severity
#Correliation matrix scikit learn library  **TRY this approach**

pd.read_sql_query("""
SELECT
    weather_severity_flag,
    AVG(dep_delayed_15) AS dep_delayed_rate,
    AVG(arr_delayed_15) AS arr_delayed_rate,
    AVG(cancelled) AS cancellation_rate
FROM joined_atl
GROUP BY weather_severity_flag;
""", conn)
```
